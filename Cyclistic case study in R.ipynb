The majority of riders opt for traditional bikes; about 8% of riders use the assistive options. Cyclistic users are more likely to ride for leisure, but about 30% use them to commute to work each day.

Until now, Cyclistic’s marketing strategy relied on building general awareness and appealing to broad consumer segments. One approach that helped make these things possible was the flexibility of its pricing plans: single-ride passes, full-day passes, and annual memberships. Customers who purchase single-ride or full-day passes are referred to as casual riders. Customers who purchase annual memberships are Cyclistic members. Cyclistic’s finance analysts have concluded that annual members are much more profitable than casual riders. Although the pricing flexibility helps Cyclistic attract more customers, The Director of Marketing Moreno believes that maximizing the number of annual members will be key to future growth. Defining Data Loading Function

```r
cmf <- function(path,pattern,output_file){
    storing_df = data.frame()
    filenames <- list.files(path, pattern=files_format, full.names=TRUE)
    for (i in 1:length(filenames)){
        if (filenames[i] != "/kaggle/input/cycling-trip-data/cyclistic_last_12_months_trips_reveresed_gps.csv"){
            temp_df <- read_csv(filenames[i])
            storing_df <- rbind(storing_df, temp_df)
    }
    }
    write.csv(storing_df, output_file ,row.names = FALSE)
    rm(storing_df)
    rm(filenames)
    rm(temp_df)
}
```

Imputating missing data with joins

In these steps, will be using the existing data to fill the missing values based on matching criteria between the fields

```r
ussi <- function(){
    swui2  <- df %>% #end station with unique Id's
        dplyr::filter(!is.na(Start_Station_Id)) %>%
        dplyr::group_by(Start_Station_Id,Start_Lat ,Start_Lng) %>%
        summarise(Count = n())

    swmi2  <- df %>% #end station with missing Id's
      dplyr::filter(is.na(Start_Station_Id)) %>%
      dplyr::group_by(Start_Lat ,Start_Lng, Start_Station_Id) %>%
      summarise(Count = n())

    found_s_id_df2 <- merge(x =swui2 , y = swmi2, by = c("Start_Lat","Start_Lng"), all.x = TRUE) %>% 
        dplyr::filter(Count.y > 0)

    found_s_id_df2 <- found_s_id_df2[, c("Start_Lat", "Start_Lng", "Start_Station_Id.x")] #stations with missing Id's found
    found_s_id_df2 <- found_s_id_df2 %>% 
      rename(Start_Station_Id_x = Start_Station_Id.x)
    
    return (sqldf(c("update df 
         set Start_Station_Id = (select Start_Station_Id_x from found_s_id_df2
         where found_s_id_df2.Start_Lat = df.Start_Lat
         and found_s_id_df2.Start_Lng = df.Start_Lng) 
         where Start_Station_Id is null", 
        "select * from df")))
    }
```

```r
ussi2 <- function(){
    swui  <- df %>% #station with unique Id's
      dplyr::filter(!is.na(Start_Station_Id) & !is.na(Start_Station_Name)) %>%
      dplyr::group_by(Start_Station_Id, Start_Station_Name) %>%
      summarise(Count = n())

    swmi  <- df %>% #station with missing Id's
      dplyr::filter(is.na(Start_Station_Id) & !is.na(Start_Station_Name)) %>%
      dplyr::group_by(Start_Station_Name, Start_Station_Id) %>%
      summarise(Count = n())

    found_s_id_df <- merge(x =swui , y = swmi, by = "Start_Station_Name", all.x = TRUE) %>% 
    dplyr::filter(Count.y > 0)

    found_s_id_df <- found_s_id_df[, c("Start_Station_Name", "Start_Station_Id.x")] #stations with missing Id's found
    found_s_id_df <- found_s_id_df %>% 
      rename(Start_Station_Id_x = Start_Station_Id.x)

    return (sqldf(c("update df 
             set Start_Station_Id = (select Start_Station_Id_x from found_s_id_df where found_s_id_df.Start_Station_Name = df.Start_Station_Name) 
             where Start_Station_Id is null", 
            "select * from df")))
}
```

```r
getSeason <- function(DATES) {
    WS <- as.Date("2012-12-15", format = "%Y-%m-%d") # Winter Solstice
    SE <- as.Date("2012-3-15",  format = "%Y-%m-%d") # Spring Equinox
    SS <- as.Date("2012-6-15",  format = "%Y-%m-%d") # Summer Solstice
    FE <- as.Date("2012-9-15",  format = "%Y-%m-%d") # Fall Equinox

    # Convert dates from any year to 2012 dates
    d <- as.Date(strftime(DATES, format="2012-%m-%d"))

    ifelse (d >= WS | d < SE, "Winter",
      ifelse (d >= SE & d < SS, "Spring",
        ifelse (d >= SS & d < FE, "Summer", "Fall")))
}
```

Reshaping data for spatial analysis

```r
Start_Loc_Member <- df %>% 
    dplyr::filter(Member_Casual =='member') %>%
    select(comm = Start_Suburb, lat = Start_Lat, lon = Start_Lng)
Start_Loc_Casual <- df %>% 
    dplyr::filter(Member_Casual =='casual') %>%
    select(comm = Start_Suburb, lat = Start_Lat, lon = Start_Lng)



End_Loc_Member <- df %>% 
    dplyr::filter(Member_Casual =='member') %>%
    select(comm = End_Suburb, lat = End_Lat, lon = End_Lng)
End_Loc_Casual <- df %>% 
    dplyr::filter(Member_Casual =='casual') %>%
    select(comm = End_Suburb, lat = End_Lat, lon = End_Lng)


Start_Points_Member = st_as_sf(Start_Loc_Member, coords = c("lon", "lat"), crs = 32616, agr = "constant")
Start_Points_Casual = st_as_sf(Start_Loc_Casual, coords = c("lon", "lat"), crs = 32616, agr = "constant")
End_Points_Member = st_as_sf(End_Loc_Member, coords = c("lon", "lat"), crs = 32616, agr = "constant")
End_Points_Casual = st_as_sf(End_Loc_Casual, coords = c("lon", "lat"), crs = 32616, agr = "constant")
```

Joining map data

```r
Start_Member_Geo <- st_join(Start_Points_Member,Chicago_File)
Start_Casual_Geo <- st_join(Start_Points_Casual,Chicago_File)
End_Member_Geo <- st_join(End_Points_Member,Chicago_File)
End_Casual_Geo <- st_join(End_Points_Casual,Chicago_File)
```

```r
Start_Member_Counts <- Start_Member_Geo %>% count(comm)
Start_Member_Counts <- Start_Member_Counts %>% rename(AGG.COUNT = n)
Start_Casual_Counts <- Start_Casual_Geo %>% count(comm)
Start_Casual_Counts <- Start_Casual_Counts %>% rename(AGG.COUNT = n)


End_Member_Counts <- End_Member_Geo %>% count(comm)
End_Member_Counts <- End_Member_Counts %>% rename(AGG.COUNT = n)
End_Casual_Counts <- End_Casual_Geo %>% count(comm)
End_Casual_Counts <- End_Casual_Counts %>% rename(AGG.COUNT = n)

Start_Member_Map_Data <- left_join(as.data.frame(Chicago_File),Start_Member_Counts,by=c("community"="comm"))
Start_Casual_Map_Data <- left_join(as.data.frame(Chicago_File),Start_Casual_Counts,by=c("community"="comm"))
End_Member_Map_Data <- left_join(as.data.frame(Chicago_File),End_Member_Counts,by=c("community"="comm"))
End_Casual_Map_Data <- left_join(as.data.frame(Chicago_File),End_Casual_Counts,by=c("community"="comm"))
```
